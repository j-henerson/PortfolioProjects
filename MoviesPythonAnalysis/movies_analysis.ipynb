```python
# Import Libraries
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts configuration of plots

# Read in the data

df = pd.read_csv(r'C:\Users\Jeremy Henerson\OneDrive\Desktop\data analysis portfolio\movies.csv')
```

```python
# Looking at the data

df.head()
```

```python
# Looking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Removing rows with NaN for Budget and Gross

df = df[df['budget'].notnull()]
df = df[df['gross'].notnull()]
```

```python
# Column data types

df.dtypes
```

```python
# Change column data types 

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# Creating a function to get the correct year from the release date, as the year released column does not always match

def getyear(date):
        return date[(date.find(',')+2):(date.find(',')+7)]
df['revised_year'] = df['released'].apply(getyear)
```

```python
df = df.sort_values(by='gross', inplace=False, ascending=False)
```

```python
# View entire dataframe
pd.set_option('display.max_rows' , None)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Scatter Plot with Budget vs. Gross 

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs. Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Plot the budget vs gross using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Observing correlation

df.corr(method='pearson')
# There is a high correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson')
# Heatmap of numeric correlations in dataset
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Looking at company
df_numeric = df

for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype == 'object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes

df_numeric.head()
```

```python
num_correlation_matrix = df_numeric.corr(method='pearson')
# Heatmap of all correlations in dataset
sns.heatmap(num_correlation_matrix, annot=True)
plt.title('Correlation Matrix for All Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
num_correlation_matrix

corr_pairs = num_correlation_matrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs > .5]

high_corr
```
